```python
import pandas as pd
from tqdm import tqdm
import numpy as np
```

```python
tr = pd.read_csv('TOTAL_TRAIN.csv')
```

```python
cols = ['edad_media', 'p_poblacion_menor_de_18',
       'p_poblacion_mayor_65', 'media_personas_por_hogar',
       'p_hogares_unipersonales', 'poblacion']
print(f"Valores ausentes antes de todo esto: {sum(tr[cols].isna().sum())}")
for col in tqdm(cols):
    for i in range(tr.shape[0]):
        try:
            float(tr[col].iloc[i])
        except:
            #print(tr[col].iloc[i])
            print(f"MOVIDA GORDA EN LA COLUMNA {col} Y LA FILA {i} ===> {tr[col].iloc[i]}")
            tr.loc[i, col] = np.nan
print(f"Valores ausentes después de todo esto: {sum(tr[cols].isna().sum())}")
```
