```python
import pickle
from lightgbm import plot_importance
```

```python
with open('best_model_INDIVIDUAL_0904_2.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
fig = plot_importance(model, ignore_zero=False, figsize=(40, 40))
```

```python
fig.figure.savefig('feature_importance_INDIVIDUAL_09.png')
```

```python
with open('X_train.pkl', 'rb') as f:
    Xtr = pickle.load(f)
```

```python
Xtr
```

```python
model.set_params(**{'reg_lambda': 10.0, 'reg_alpha': 0.0067, 'n_estimators': 3500, 'learning_rate': 0.03, 'num_leaves': 60})
```

```python
with open('y_train.pkl', 'rb') as f:
    y_train = pickle.load(f)
```

```python
i = 0
good_colnames = []
for col in Xtr.columns:
    if not col.isascii():
        good_colnames.append(f'{i}_not_ascii')
    else:
        good_colnames.append(col)
    i += 1
```

```python
Xtr.columns = good_colnames
```

```python
model.fit(Xtr, y_train)
```

```python
with open('X_test.pkl', 'rb') as f:
    Xtest=pickle.load(f)
```

```python
with open('y_test.pkl', 'rb') as f:
    y_test=pickle.load(f)
```

```python
from sklearn.metrics import f1_score
```

```python
Xtest.columns = good_colnames
```

```python
f1_score(y_test, model.predict(Xtest), average='macro')
```
