```python
def probar_a_transformar(X, Y):
    from tqdm import tqdm
    for i in tqdm(range(60000, 150000)):
        try:
            trans = []
            inProj = Proj(init=f'epsg:{i}')#Proj(init='epsg:25830') #2805
            outProj = Proj(init='epsg:4326')
            for x, y in zip(X, Y):
                trans.append(transform(inProj, outProj, x, y))
            if trans[5][0] > -5. and trans[5][0] <= -1. and trans[5][1] > 35 and trans[5][1] <= 48:
                print(f'El código {i} podría ser el bueno, ahí van las transformadas')
                return trans
        except:
            continue
```
