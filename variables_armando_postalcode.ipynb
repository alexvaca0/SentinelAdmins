```python
def get_madrid_codes(df):
    
    df.dropna(inplace=True)
    #df.CODIGO_POSTAL = df.CODIGO_POSTAL.astype('str')
    #df_madrid = df[df.CODIGO_POSTAL.find('28')]
    df_madrid = df[(df.CODIGO_POSTAL>27999) & (df.CODIGO_POSTAL<29000)]
    df_madrid.CODIGO_POSTAL = df.CODIGO_POSTAL.astype('int')
    return df_madrid
```

```python
def get_lon_lat_postal_codes(df):
    nomi = pgeocode.Nominatim('es')
    unique_codes = df.CODIGO_POSTAL.unique()
    dic_ = {'CODIGO_POSTAL': [c for c in unique_codes],
            'lon': [],
            'lat': []}
    for code in dic_['CODIGO_POSTAL']:
        res = nomi.query_postal_code(f"{code}")
        dic_['lon'].append(res['longitude'])
        dic_['lat'].append(res['latitude'])
    
    df_lon_lat = pd.DataFrame(dic_)
    df_lon_lat.CODIGO_POSTAL = df_lon_lat.CODIGO_POSTAL.astype('int')
    df_total = pd.merge(df, df_lon_lat, how='outer', on='CODIGO_POSTAL')
    return df_total
    
    
```
