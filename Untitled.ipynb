```python
lat, lon = [], []

for index, row in gdf.iterrows():
    try:
        for pt in list(row['geometry'].exterior.coords): 
            lat.append(pt[0])
            lon.append(pt[1])
    except Exception as e:
        print(e)
```

```python
myProj = Proj("+proj=utm +zone=89, +north +ellps=WGS84 +datum=WGS84 +units=m +no_defs")
```

```python
lat, lon = myProj(lat,lon, inverse=True)
```

```python
from pyproj import Proj, transform

inProj = Proj(init='epsg:25830')
outProj = Proj(init='epsg:4326')
```

```python
latnew, lonnew = [], []
for la, lo in zip(lat, lon):
    a, b = transform(inProj,outProj,la, lo)
    latnew.append(a)
    lonnew.append(b)
```
